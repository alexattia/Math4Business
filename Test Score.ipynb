## Etherscan

```python
df = parse_ether(100)
```

```python
df.head()
```

```python
df['Average Reward fee per transac'] = df['Reward'].mean()/df['txn']

df['datetime'] = pd.to_datetime(df['Age'], format="%b-%d-%Y %H:%M:%S %p")
df['time_block'] = df['datetime'].diff(periods=-1)

df.txn.mean()
df['Average Reward fee per transac'].mean()
df['Reward'].mean()
df['txn'].mean()
df['time_block'].mean()

df['Average Reward fee per transac'].std()
df['Reward'].std()
df['txn'].std()
df['time_block'].std()

df['Average Reward fee per transac'].quantile(0.95)
df['Reward'].quantile(0.95)
df['txn'].quantile(0.05)
df['time_block'].quantile(0.95)
```
