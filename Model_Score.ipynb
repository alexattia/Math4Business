```python
#Computing intermediate non-normalized scores
Model.loc['Liquidity_'] = (np.power(Model.loc['Market Capitalization (in USD)'],0.1)*Model.loc['Avg Nb transac']*0.000001)/Model.loc['Avg Waiting time between two blocs']
Model.loc['Cost_'] = np.abs(1 - Model.loc['Percentile 95% fee per transaction']/Model.loc['Avg fee per transaction'])*Model.loc['Price (in USD)']*(Model.loc['Avg fee per transaction']+0.4*Model.loc['St dev fee per transaction'])
Model.loc['Processing time_'] = np.log((Model.loc['St dev waiting time between 2 blocs']/(Model.loc['Avg Waiting time between two blocs']*Model.loc['Percentile 95% waiting time']))**3)
```

```python
Model
```

```python
#Normalizing intermediate scores
Liquidiy_norm = np.max(Model.loc['Liquidity_'])
Cost_norm = np.sum(Model.loc['Cost_'])
Time_norm = np.max(Model.loc['Processing time_'])
print(Time_norm)

Model.loc['Liquidity'] = Model.loc['Liquidity_']/Liquidiy_norm
Model.loc['Cost'] = 1 - Model.loc['Cost_']/Cost_norm
Model.loc['Processing time'] = Model.loc['Processing time_']/Time_norm
```

```python
Model.loc['Score'] = (W[0]*Model.loc['Liquidity']+W[1]*Model.loc['Cost']+W[2]*Model.loc['Processing time'])/np.sum(W)
```
